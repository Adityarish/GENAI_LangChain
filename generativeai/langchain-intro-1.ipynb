### Streaming and batch processing
#### Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, especially for long-running tasks.Calling stream() on the model returns a generator that yields the model's output as it is generated.

```python
model.invoke("Write a note on reproduction in 300 words")
```

```python
for chunk in model.stream("Write a note on reproduction in 300 words"):
    print(chunk.content, end="|", flush=True)
```

## Batch processing
##### Batch is collection of independent requests to a model can significanlty imporve performance can reduce costs , as the procesing can be done in parallel

```python
response = model.batch(
    [
        "Write a note on reproduction in 100 words",
        "Write a note on photosynthesis in 100 words",
    ],
    config={
        'max_concurrency': 5, # it means max 5 requests can be sent at a time
    }
)
for res in response:
    print(res.content)
```
